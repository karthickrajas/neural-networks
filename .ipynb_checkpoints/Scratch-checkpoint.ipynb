```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
im = cv2.imread("karthick.png")
print(im.shape)
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
gray = rgb2gray(im)    
plt.imshow(gray,cmap = plt.get_cmap('gray'))
plt.show()
```

```python
gray.shape
```

```python
kernel1= np.array([[-1, -1, -1], 
                   [1, 1, 1],
                   [-1, -1, -1]])
```

```python
def apply_kernel(image,kernel):
    result = np.zeros(image.shape)

    # Output array
    for ii in range(image.shape[0] - kernel.shape[0]):
        for jj in range(image.shape[1] - kernel.shape[0]):
            result[ii, jj] = (image[ii:ii+3, jj:jj+3] * kernel).sum()

    # Print result
    return(result)
```

```python
plt.imshow(apply_kernel(gray, kernel1))
plt.show()
```

```python
kernel2 = np.array([[1, 1, 1], 
                   [1, 0, 1],
                   [1, 1, 1]]) # vertical lines
```

```python
plt.imshow(apply_kernel(gray, kernel2))
plt.show()
```

```python
def max_pool(image):
    # Result placeholder
    pool = np.zeros((image.shape[0]//2, image.shape[1]//2))

    # Pooling operation
    
    for ii in range(pool.shape[0]):
        for jj in range(pool.shape[1]):
            pool[ii, jj] = np.max(image[ii*2:ii*2+2,jj*2:jj*2+2])
            
    return(pool)
```

```python
plt.imshow(apply_kernel(max_pool(max_pool(max_pool(max_pool(gray)))),kernel2),cmap='gray')
plt.show()
```

```python
max_pool(max_pool(max_pool(gray))).shape
```

```python
plt.imshow(apply_kernel(max_pool(max_pool(max_pool(max_pool(gray)))),kernel1),cmap='gray')
plt.show()
```

```python
reduced_image = max_pool(max_pool(max_pool(max_pool(gray))))
```

```python
plt.hist(reduced_image.ravel(),bins=50)
plt.show()
```
